## 다문화 가정 시설, 인구시각화, 강서구와 서울 전체 평균

### 1. 다문화 가정 인구 시각화

```python
import pandas as pd
import matplotlib as mp
```

```python
df=pd.read_csv('다문화가정데이터.csv',thousands=',')
df
```

```python
#결측값 처리
df1=df.fillna('2020')
```

```python
df1.info()
```

```python
import matplotlib.pyplot as plt
import matplotlib
import seaborn as sns
matplotlib.rcParams['font.family']='AppleGothic'
matplotlib.rcParams['font.size']=10
matplotlib.rcParams['axes.unicode_minus']=False
```

```python
df2=df1.drop(index=0,axis=0)
df2
```

## 데이터 시각화

```python
df3 = df2.sort_values('다문화가구', ascending=False)
plt.figure(figsize = (16,8))

sns.lineplot(
    x = '자치구',
    y = '다문화가구',
    data = df3,
)
#서울 전체에서 구로구, 영등포구, 금천구, 관악구, 강서구의 다문화 가구 인구수가 차례로 많음
```

```python
#내국인, 외국인 별 시각화
fig, ax = plt.subplots()
fig.subplots_adjust(hspace=0.1, wspace=0.1, left = 34, right = 40)
for idx in range(3):
    plt.subplot(1,3, idx+1)
    fam_df = df2.groupby("자치구")[list(df2.columns[-3:])[idx]].describe()
    sns.barplot(fam_df.index, fam_df.loc[:,"mean"].values)
    plt.ylabel("score")
    plt.title(list(df2.columns[-3:])[idx])
    
plt.show()
```

```python
import matplotlib.pyplot as plt
import matplotlib
matplotlib.rcParams['font.family']='AppleGothic'
matplotlib.rcParams['font.size']=10
matplotlib.rcParams['axes.unicode_minus']=False
```

```python
plt.figure(figsize = (15,8))
sns.barplot(x = df2['자치구'].value_counts().index,
           y= df2['다문화가구'], palette = 'terrain')
plt.xlabel('자치구')
plt.ylabel('다문화가구')
plt.show()
```

```python
import matplotlib.pyplot as plt
df3=df2['다문화가구'].head(5)
plt.pie(df3)
plt.title('다문화 가구 수가 많은 상위 5개 자치구')
plt.pie(df3,startangle=90,explode=(0,0,0,0,0.1),autopct='%1.1f%%')
plt.legend(labels=df2['자치구'],loc='lower left',bbox_to_anchor=(0.9,0.1))
```

```python
#서울시 전체 평균 다문화 가구 수와 강서구 다문화 가구 수 비교
#서울시 전체 평균 : 2701.6
#강서구 : 3766

#서울시 다문화 가정 총계 : 187,239
#서울시 다문화 가정 평균 : 7489.56
#강서구 다문화 가정 총계 : 10,421

#강서구는 서울시 전체 다문화 가구 수보다 많은 인구가 살고 있는 축에 속하고 총합의 평균을 비교해봤을 때도 상위 5위 자치구이다.
```
