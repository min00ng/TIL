### 1. 다문화 가정 인구 시각화

```python
import pandas as pd
import matplotlib as mp
```

```python
df=pd.read_csv('다문화가정데이터.csv',thousands=',')
df
```

```python
#결측값 처리
df1=df.fillna('2020')
```

```python
df1.info()
```

```python
import matplotlib.pyplot as plt
import matplotlib
import seaborn as sns
matplotlib.rcParams['font.family']='AppleGothic'
matplotlib.rcParams['font.size']=10
matplotlib.rcParams['axes.unicode_minus']=False
```

```python
df2=df1.drop(index=0,axis=0)
df2
```

## 데이터 시각화

```python
df3 = df2.sort_values('다문화가구', ascending=False)
plt.figure(figsize = (16,8))

sns.lineplot(
    x = '자치구',
    y = '다문화가구',
    data = df3,
)
#서울 전체에서 구로구, 영등포구, 금천구, 관악구, 강서구의 다문화 가구 인구수가 차례로 많음
```

```python
#내국인, 외국인 별 시각화
fig, ax = plt.subplots()
fig.subplots_adjust(hspace=0.1, wspace=0.1, left = 99, right = 105)
for idx in range(3):
    plt.subplot(1,3, idx+1)
    fam_df = df2.groupby("자치구")[list(df2.columns[-3:])[idx]].describe()
    sns.barplot(fam_df.index, fam_df.loc[:,"mean"].values)
    plt.ylabel("score")
    plt.title(list(df2.columns[-3:])[idx])
    
plt.show()
```

```python
#서울시 전체 평균 다문화 가구 수와 강서구 다문화 가구 수 비교
#서울시 전체 평균 : 2701.6
#강서구 : 3766

import matplotlib.pyplot as plt
import numpy as np

x = np.arange(2)
seoul = ['서울시 평균 다문화 가구 수', '강서구 다문화 가구 수']
values = [2701.6, 3766]

plt.bar(x, values,color=['pink','y'])
plt.xticks(x,seoul)

plt.show()
```

## 2. 다문화 가정 시설

```python
fam=pd.read_csv('다문화시설.csv')
fam
#각 자치구마다 하나씩 다문화 가족 지원센터가 있음
```

```python
from folium.map import Popup
import pandas as pd
import folium

df=pd.read_csv('다문화시설.csv')
seoul_map=folium.Map(location=[37.55,127],titles='Stamen Terrain',
                    zoom_start=11)
#위치 정보를 Marker 로 표시
#index를 시설명으로 변경하고 시설 이름이 표시되도록 함
for name, lat, lng in zip(df.시설명,df.위도,df.경도):
    iframe=folium.IFrame(name,width=100,height=80)
    popup=folium.Popup(iframe,max_width=500)
    folium.Marker([lat,lng],popup=popup).add_to(seoul_map)
    
seoul_map
```

```python
#강서구는 서울시 중 다문화 가구 수가 많은 자치구 상위 5개에 속한다.
#최근 유입되고 있는 많은 외국인 노동자와 늘어나는 다문화 가정을 위해 다문화 가정 센터 연결하여 한국에서의 생활을 원활하게 한다.
```

## 서울시 다문화 가정 학생 데이터

```python
stu=pd.read_csv('다문화가정학생.csv',thousands=',')
stu
```

```python
stu1=stu.fillna(2021)
stu2=stu1.drop(index=0,axis=0)
stu2
```

```python
#두가지의 시사점
#초등학교 다문화 가정 아이들은 10000명, 고등학교는 5분의 1인 2000명 정도
#고등학교 때까지 한국 학교에서 적응을 하지 못해 중고등학교에 대부분 포기한다는 가정
#전체적으로 다문화 가정이 늘어나서 초등학생 아이가 있는 다문화 가정의 비율이 가장 많다는 가정
```

```python
stu3 = stu2.sort_values('초등학교 총계', ascending=False)
plt.figure(figsize = (16,8))

sns.lineplot(
    x = '자치구',
    y = '초등학교 총계',
    data = stu3,
)
#초등학교 학생의 경우 강서구가 4위 
```

```python
stu4 = stu2.sort_values('중학교 총계', ascending=False)
plt.figure(figsize = (16,8))

sns.lineplot(
    x = '자치구',
    y = '중학교 총계',
    data = stu4,
)
#중학교의 경우 강서구 학생들의 이탈률이 많은 것으로 보인다
```

```python
stu5 = stu2.sort_values('고등학교 총계', ascending=False)
plt.figure(figsize = (16,8))

sns.lineplot(
    x = '자치구',
    y = '고등학교 총계',
    data = stu5,
)
```
